```python
#Group the communes by canton and save them into separate .csv files, e.g. a ZH.csv with all the data for communes in Zurich (Do not include the .csv files in your submission).(commande from ChatGPT)

# Group by 'canton' (replace 'canton' with your actual column name)
grouped = df.groupby('Canton')

# Save each group to a separate CSV file
for name, group in grouped:
    group.to_csv(f'{name}.csv', index=False)
```

```python
#B5 Compute the population density at the canton level and rank the cantons from most dense to least dense. Clearly state and comment your observations.(commande from ChatGPT)
# Group by 'Canton' and compute mean population density
mean_density_by_canton = df.groupby('Canton')['Population density per km²'].mean()

# Sort by population density (most dense to least dense)
ranked_cantons = mean_density_by_canton.sort_values(ascending=False)

# Display the result
ranked_cantons
```

```python
#B6 Compute the number of communes in each canton where more than 50 percent of their populations are aged between 20 and 64 years old.
```

```python
# Calculate population percentage (20 to 64 years old)
df['Population Percentage'] = (df['Age20_64'] / df['Population']) * 100

# Filter communes where population percentage > 50%
filtered_df = df[df['Population Percentage'] > 50]

# Group by 'Canton' and count communes
communes_per_canton = filtered_df.groupby('Canton')['Commune'].count()

# Display the result
print(communes_per_canton)
```
